```python
#clean and reformat 
#observing unique values, there are no negative values in all numerical cols. Only Nett price has '-' which needs to covert to null

df = pd.read_csv("landed_all.csv") 
df['No. of Units'] = df['No. of Units'].astype('int')

df['Area (sqm)'] = df['Area (sqm)'].str.replace(',', '')
df['Area (sqm)'] = df['Area (sqm)'].astype('float')

df['Transacted Price ($)'] = df['Transacted Price ($)'].str.replace(',', '')
df['Transacted Price ($)'] = df['Transacted Price ($)'].astype("float")

df['Nett Price($))'] = df['Nett Price($)'].str.replace(',', '')
df['Nett Price($))'] = df['Nett Price($))'].replace('-', np.NaN)

df['Unit Price ($ psm)'] = df['Unit Price ($ psm)'].str.replace(',', '')
df['Unit Price ($ psm)'] = df['Unit Price ($ psm)'].astype("float")


df['Unit Price ($ psf)'] = df['Unit Price ($ psf)'].str.replace(',', '')
df['Unit Price ($ psf)'] = df['Unit Price ($ psf)'].astype("float")


#sale date Hi TAs, error here
#df['Sale Date']=(df['Sale Date'].astype('str')).astype('datetime64[ns]')
#df['Sale Date'] = df['Sale Date'].astype('datetime64[ns]')
for i in (df['Sale Date'].unique()):
    print (i)
    

#change type to datetime64[ns]
df['Completion Date'] = df['Completion Date'].replace('-', np.NaN)
df['Completion Date'] = df['Completion Date'].replace('Unknown', np.NaN)
df['Completion Date'] = df['Completion Date'].replace('Uncompleted', np.NaN)
#df['Completion Date']=(df['Completion Date'].astype('str')).astype('datetime64[ns]')

#tenure, find a way change to expiry date 

df['Purchaser Address Indicator'] = df['Purchaser Address Indicator'].replace('N.A', np.NaN)

```
